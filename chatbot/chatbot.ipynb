```python
# Sample users data
users = [
    {"id":1, "name":"Student1", "courses":["Machine Learning", "Calculus II"], "role":"student"},
    {"id":2, "name":"Alice", "courses":["Machine Learning", "Physics I"], "role":"topper"},
    {"id":3, "name":"Bob", "courses":["Calculus II", "Chemistry I"], "role":"topper"},
    {"id":4, "name":"Charlie", "courses":["Machine Learning", "Biology I"], "role":"topper"}
]

# Mini-chatbot function: step 1 only
def mini_chatbot_find_users(message):
    message = message.lower()
    
    for course in [c.lower() for c in sum([u["courses"] for u in users], [])]:
        if course in message:
            matched = [u["name"] for u in users if course in [c.lower() for c in u["courses"]]]
            return f"Here are users for {course.title()}: {', '.join(matched)}"
    
    return "Sorry, no users found for that subject."

# Example usage
print(mini_chatbot_find_users("I want students of Machine Learning"))
print(mini_chatbot_find_users("Show me toppers for Calculus II"))
```

```python
# Example messages
messages = [
    "I want students of Machine Learning",
    "Show me toppers for Calculus II",
    "Do you have students for Physics I?",
    "Find me users of Biology I",
    "Is there anyone for Chemistry I?",
    "I want students for English"  # not in the dataset
]

# Loop through messages and get chatbot responses
for msg in messages:
    response = mini_chatbot_find_users(msg)
    print(f"Message: {msg}")
    print(f"Chatbot: {response}\n")
```
